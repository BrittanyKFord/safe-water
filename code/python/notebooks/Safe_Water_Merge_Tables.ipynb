```python
cmd = "select PWSID, SERVICE_AREA_TYPE_CODE from SERVICE_AREA"
serv_area = pd.read_sql(cmd, db)
serv_area.head()
```

```python
cmd = "select PWSID, FACILITY_ID, TREATMENT_OBJECTIVE_CODE, TREATMENT_PROCESS_CODE from TREATMENT"
treat = pd.read_sql(cmd, db)
treat.head()
```

```python
def get_year_for_violation(row):
    current_year = np.datetime64('now').astype(object).year
    if row['COMPL_PER_BEGIN_DATE'] is not None:
        return str(current_year - row['COMPL_PER_BEGIN_DATE'].year) + '_yr_ago'
    else:
        return 'unknown_year'
```

```python
cmd = "select PWSID, FACILITY_ID, COMPL_PER_BEGIN_DATE, CONTAMINANT_CODE, CORRECTIVE_ACTION_ID, " \
    + "IS_HEALTH_BASED_IND, IS_MAJOR_VIOL_IND, LATEST_ENFORCEMENT_ID, PUBLIC_NOTIFICATION_TIER, " \
    + "RTC_DATE, RTC_ENFORCEMENT_ID, RULE_CODE, RULE_FAMILY_CODE, RULE_GROUP_CODE, SEVERITY_IND_CNT, " \
    + "VIOLATION_CATEGORY_CODE, VIOLATION_CODE from VIOLATION " \
    + "where IS_HEALTH_BASED_IND = 'Y' " \
    + "and COMPL_PER_BEGIN_DATE >= '2009-01-01' " \
    + "and COMPL_PER_BEGIN_DATE <= '2016-12-31'"
viol = pd.read_sql(cmd, db, parse_dates=['COMPL_PER_BEGIN_DATE'])
viol.loc[:, 'VIOLATION_YEARS_AGO'] = viol.apply(get_year_for_violation, axis=1)
viol.head()
```
